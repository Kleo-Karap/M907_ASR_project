# Plotting the WER results from the generated .csv files

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Original data

```python
wer_results_clean=pd.read_csv("results_clean.csv", sep=',', decimal=".")
wer_results_clean.head()
```

```python
# Plotting the WER values for each sentence
plt.figure(figsize=(10, 6))
plt.plot(wer_results_clean['WAV_ID'], wer_results_clean['WER'])
plt.title('Word Error Rate per Sentence in original files')
plt.xlabel('WAV_ID')
plt.ylabel('WER')
# Remove individual numbers from x-axis labels , because wav-id numbers take too much space
plt.gca().xaxis.set_ticklabels([''] * len(wer_results_clean.index))
#plt.xticks(rotation=90)
plt.show()
```

# Getting the mean WER from each csv file to generate a comparative plot

```python
#Getting mean values of each dataframe  after removing 1 outlier (8842-304647-0001_3), which as we discussed earlier, is a
#non-English sentence and has extreme WER value.
mean_wer_clean=new_clean['WER'].mean()
mean_wer_clean_lm_0=new_clean_lm_0['WER'].mean()
mean_wer_clean_lm_1=new_clean_lm_1['WER'].mean()
mean_wer_clean_lm_1_5=new_clean_lm_1_5['WER'].mean()
mean_wer_clean_lm_2=new_clean_lm_2['WER'].mean()
mean_wer_noisy_3 = new_noisy_3['WER'].mean()
mean_wer_noisy_6 = new_noisy_6['WER'].mean()
mean_wer_noisy_9 = new_noisy_9['WER'].mean()
mean_wer_clean_lm_beta=new_clean_lm_beta['WER'].mean()

print(mean_wer_clean_lm_beta)
print(mean_wer_clean)
```

```python
import matplotlib.pyplot as plt

# Dataset names
dataset_names = ['Clean','Clean_LM Beta','Clean_LM 0', 'Clean_LM 1', 'Clean_LM 1.5', 'Clean_LM 2', 'Noisy 3', 'Noisy 6', 'Noisy 9']
dataset_values = [mean_wer_clean,mean_wer_clean_lm_beta, mean_wer_clean_lm_0, mean_wer_clean_lm_1, mean_wer_clean_lm_1_5, mean_wer_clean_lm_2, mean_wer_noisy_3, mean_wer_noisy_6, mean_wer_noisy_9]
x_values = range(1, len(dataset_values) + 1)
plt.plot(x_values, dataset_values, marker='o')
plt.title("Mean WER Values of Datasets")
plt.xlabel("Dataset")
plt.ylabel("Mean WER")
plt.xticks(x_values, dataset_names, rotation=45)
plt.show()
```

From the above plot we can notice that the weight of the language model does have an impact on the model's performance. In fact, when lm_alpha vlaue is 0,1, or 1.5(twice the default value) we observe similar WER results (17%), but still twice higher compared to the clean files (8%) (where lm_alpha=0.75). It's is obvious though, that when we exceed that limit and set the lm_alpha=2, the WER reaches the 30%! We can say, that with that weight on the language model, the acoustic model does not exist.
We rely only on the language model probabilities for the final transcription.This behavior can be described as "overfitting" to the language model.


Comparing the results of the noisy files, we confirm our intuition about the model's performance increase with higher SNR. We see that an SNR of 3dB gives us the highest WER result (over 35%). But as the SNR increases, the WER decreases. The noisy file with SNR value of 9dB gives a WER result quite close to the one from the clean files with adjusted language model weights. So could say that adding noise with SNR=9dB will not have significant effect on the original signal.
